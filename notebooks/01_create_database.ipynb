```python
import pandas as pd
import sqlite3

df = pd.read_csv("C:/Users/aidan/Downloads/housing_app_fall25-main/housing_app_fall25-main/Telco_Customer_Churn.csv")
```

Clean + Convert 

```python
df.columns = df.columns.str.strip()
df["Churn"] = (df["Churn"] == "Yes").astype(int)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
```

```python
Database Connections
```

```python
conn = sqlite3.connect("project.db")
cur = conn.cursor()
```

```python
cur.executescript("""
-- SQL CREATE TABLE statements
""")
conn.commit()
```

```python
# ---- Customer table ----
df[[
    "customerID", "gender", "SeniorCitizen", "Partner", "Dependents"
]].rename(columns={
    "customerID": "customer_id",
    "SeniorCitizen": "senior_citizen"
}).to_sql(
    "Customer", conn, if_exists="append", index=False
)

# ---- Services table ----
df[[
    "customerID", "PhoneService", "MultipleLines", "InternetService",
    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies"
]].rename(columns={
    "customerID": "customer_id",
    "PhoneService": "phone_service",
    "MultipleLines": "multiple_lines",
    "InternetService": "internet_service",
    "OnlineSecurity": "online_security",
    "OnlineBackup": "online_backup",
    "DeviceProtection": "device_protection",
    "TechSupport": "tech_support",
    "StreamingTV": "streaming_tv",
    "StreamingMovies": "streaming_movies"
}).to_sql(
    "Services", conn, if_exists="append", index=False
)

# ---- Contract table ----
df[[
    "customerID", "Contract", "PaperlessBilling", "PaymentMethod"
]].rename(columns={
    "customerID": "customer_id",
    "Contract": "contract",
    "PaperlessBilling": "paperless_billing",
    "PaymentMethod": "payment_method"
}).to_sql(
    "Contract", conn, if_exists="append", index=False
)

# ---- Billing table ----
df[[
    "customerID", "tenure", "MonthlyCharges", "TotalCharges"
]].rename(columns={
    "customerID": "customer_id",
    "MonthlyCharges": "monthly_charges",
    "TotalCharges": "total_charges"
}).to_sql(
    "Billing", conn, if_exists="append", index=False
)

# ---- Churn table (target) ----
df[[
    "customerID", "Churn"
]].rename(columns={
    "customerID": "customer_id",
    "Churn": "churn"
}).to_sql(
    "Churn", conn, if_exists="append", index=False
)
```
